2. วาด boxplot เปรียบเทียบ การกระจายตัวของข้อมูล PetalLength ของดอก iris ทั้ง 3 ชนิด

1. เขียน function ที่รับ input เป็น output ของ boxplot เเล้ว แสดงค่า min,max,q1,q2,q3

```python
x = plt.boxplot(df['SepalWidth'])
```

```python
def box_vals(x):
    q1 = x['boxes'][0].get_ydata()[0]
    q2 = x['medians'][0].get_ydata()[0]
    q3 = x['boxes'][0].get_ydata()[2]
    iqr = q3-q1
    vmin = min(x['whiskers'][0].get_ydata())
    vmax = max(x['whiskers'][1].get_ydata())
    return  print(f'Q1 เท่ากับ {q1}\nQ2 เท่ากับ {q2}\nQ3 เท่ากับ {q3}\nIQR เท่ากับ {iqr}\nMin เท่ากับ {vmin}\nMax เท่ากับ {vmax}')
```

```python
box_vals(x)
```

2. วาด boxplot เปรียบเทียบ การกระจายตัวของข้อมูล PetalLength ของดอก iris ทั้ง 3 ชนิด

```python
#กำหนดขนาดของกราฟ 
plt.figure(figsize =(10,5))  # (กว้าง,ยาว)

#plot กราฟ Iris-setosa
plt.subplot(1,3,1)  # กำหนดตำแหน่งกราฟ (แถว,คอลัมน์,ตำแหน่ง)
plt.boxplot(df['PetalLength'][0:50],vert=True,meanline=True,showmeans=True);  #Iris-setosa (boxplot อันที่ 1)
plt.title('Iris-setosa');  # ตั้งชื่อกราฟ
plt.ylabel('PetalLength');  # ตั้งชื่อแกน y

#plot กราฟ Iris-versicolor
plt.subplot(1,3,2)  #กำหนดตำแหน่งกราฟ (แถว,คอลัมน์,ตำแหน่ง)
versicolor_bp = plt.boxplot(df['PetalLength'][50:100],vert=True,meanline=True,showmeans=True);  #Iris-versicolor (boxplot อันที่ 2)
plt.title('Iris-versicolor');  #ตั้งชื่อกราฟ
plt.ylabel('PetalLength');  #ตั้งชื่อแกน y

#plot กราฟ Iris-virginica
plt.subplot(1,3,3)  # กำหนดตำแหน่งกราฟ (แถว,คอลัมน์,ตำแหน่ง)
virginica_bp = plt.boxplot(df['PetalLength'][100:150],vert=True,meanline=True,showmeans=True);  #Iris-virginica (boxplot อันที่ 3)
plt.title('Iris-virginica');  #ตั้งชื่อกราฟ
plt.ylabel('PetalLength');  #ตั้งชื่อแกน y

plt.tight_layout()  #ปรับระยะห่างของแต่ละกราฟ
```
